```python
from collections import Counter
```

```python
def stats(X):
    C = Counter({
        "n_nodes": 1,
        "n_women": int( X['gender'] == 'woman' ),
        "n_men": int( X['gender'] == 'man' ),
    })
    
    if 'children' in X:
        for c in X['children']:
            C += stats(c)
            
    return C
```

```python
def stats(X):
    C = Counter({
        "n_nodes": 1,
        "n_women": int( X['gender'] == 'woman' ),
        "n_men": int( X['gender'] == 'man' ),
        "n_other": int( X['gender'] not in {'man','woman'} ),
    })
    #print(X['gender'])
    
    if 'children' in X:
        for c in X['children']:
            C += stats(c)
            
    return C
```

```python
import json
for fn in fns:
    print(fn.name)
    with fn.open() as inf:
        tr = json.loads(inf.read())
        
        C = stats(tr)
        C['pct_women'] = f"{C['n_women'] / C['n_nodes']:0.1%}"
        
        for k,v in sorted(C.items()):
            print(f"\t{k}: {v}")
```

```python
def print_tree(X, depth=0, maxdepth=3):
    if depth >= maxdepth:
        return
    
    s = stats(X)
    print( "\t"*depth + X['name'].title() + f" ({s['n_nodes']}, {s['n_women']/s['n_nodes']:0.1%}F)" )
    
    if 'children' in X:
        for c in X['children']:
            print_tree(c, depth=depth+1)
            
```

```python
import json
for fn in fns:
    print(fn.name)
    with fn.open() as inf:
        tr = json.loads(inf.read())
        print_tree(tr)
    print()
```
